```python
import random
import string
import nltk # Corrected 'nitk' to 'nltk'

from nltk.corpus import words

nltk.download('words')

word_list = words.words() # Fixed assignment operator

def generate_random_word():
    return random.choice(word_list)

print('generating 5 words')
for _ in range(5): # Changed 'in' to '_' for unused loop variable
    print(generate_random_word())
```
